```python
from model_lstm import LSTMModel
import pandas as pd
import numpy as np
```

```python
#load the full training daataset. this is just to fit a tokenizer
df = pd.read_csv('datasets/train_data.csv')

print(df.head())
print(df["SentimentText"].to_numpy())
```

```python
import nltk
# nltk.download('wordnet')
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load spaCy English model
# lemmatizer = nltk.stem.WordNetLemmatizer()

# Example texts
texts = ["text sample 1", "text sample 2", "text sample 3"]

# Lemmatization function using spaCy
# def lemmatize_text(text):
#     doc = lemmatizer.lemmatize(text)
#     return doc

# Apply lemmatization to each text in the dataset
# lemmatized_texts = [lemmatize_text(text) for text in texts]

# Tokenization using Keras Tokenizer
vocab_size = 10000
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>", lower=True)
# tokenizer.fit_on_texts(lemmatized_texts)

tokenizer.fit_on_texts(df["SentimentText"].to_numpy())

tokenizer.vocab_size = vocab_size
# Convert text to sequences
# sequences = tokenizer.texts_to_sequences(lemmatized_texts)
sequences = tokenizer.texts_to_sequences(df["SentimentText"].to_numpy())

# Pad sequences
max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len)

# Vocabulary size
vocabulary_size = tokenizer.num_words + 1
print("Vocabulary Size:", vocabulary_size)
```
