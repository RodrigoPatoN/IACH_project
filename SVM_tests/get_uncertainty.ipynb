## Active learning data selection

```python
def get_most_uncertain_rows(dataset, model, n_rows=1000):

    dataset = deepcopy(dataset)

    distance_to_margin = model.decision_function(dataset)
    absolute_distance = np.abs(distance_to_margin)
    
    dataset["absolute_distance_to_margin"] = absolute_distance
    
    result_df_sorted = dataset.sort_values(by='absolute_distance_to_margin', ascending=True)

    result_df_sorted = result_df_sorted.drop(columns=['absolute_distance_to_margin'])
        
    return result_df_sorted.head(n_rows)


def compute_metrics(y_true, y_pred):
    
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    return accuracy, precision, recall, f1
```
